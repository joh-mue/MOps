```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
%matplotlib inline
x = [a for a in range(1000)]
naive_x = [a for a in range(1000)]
naive = [a**3 for a in range(1000)]
strassen = [a**2.8074 for a in range(1000)]
best = [a**2.374 for a in range(1000)]

matplotlib.rcParams.update({'font.size': 12})
matplotlib.rcParams.update({'figure.autolayout': True})

fig, ax = plt.subplots()
ax.plot(naive_x, naive, color='red', linestyle='-', linewidth=2)
ax.plot(x, strassen, color='blue', linestyle='-', linewidth=2)
ax.plot(x, best, color='green', linestyle='-', linewidth=2)

ax.legend(('naive', 'strassen', 'best'), loc=2)
ax.grid()
ax.set_xlabel('Elements n')
ax.set_ylabel('Number of Operations in billions.')

plt.savefig('/Users/Johannes/Dropbox/thesis/img/naive-strassen.png', dpi=600)
```
